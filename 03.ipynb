```python
import numpy as np
```

```python
with open("03.txt", "r") as f:
    txt = f.readlines()
banks = [[int(l) for l in list(line.rstrip())] for line in txt]
```

```python
# 1
result = 0
for bank in banks:
    p1 = np.argmax(bank[:-1])
    p2 = np.argmax(bank[p1+1:]) + p1 + 1 
    
    num = bank[p1] * 10 + bank[p2]
    result += num
print(result)
```

```python
# 2
result = 0
for bank in banks:
    pos = 0
    num = 0
    for i in range(12):
        p = np.argmax(bank[pos:-(12-i)])
        pos = p + 1
        num = num * 10 + bank[p]
        
    result += num
    print(num)
```
