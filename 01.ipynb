```python
import numpy as np
```

```python
with open("01.txt", "r") as f:
    txt = f.readlines()
instr = [(line[0], int(line[1:-1])) for line in txt]
```

```python
# 1
pos = 50
password = 0
for (d, c) in instr:
    pos = (pos + c if d == 'R' else pos - c) % 100    
    if pos == 0:
        password += 1
print(password)
    
```

```python
# 2
pos = 50
password = 0
for (d, c) in instr:
    for _ in range(c):
        pos = (pos + 1 if d == 'R' else pos - 1) % 100
        if pos == 0:
            password += 1

print(password)
    
```
