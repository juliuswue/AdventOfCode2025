```python
import numpy as np
```

```python
with open("02.txt", "r") as f:
    txt = f.readlines()[0].split(',')
ids = [x.split('-') for x in txt]
```

```python
# 1
result = 0
for identifiers in ids:
    for no in range(int(identifiers[0]), int(identifiers[1])+1):
        l = len(str(no))
        no_str = str(no)
        if l % 2 == 0:
            if no_str[:l//2] == no_str[l//2:]:
                result += no
print(result)
```

```python
# 2
result = 0
for identifiers in ids:
    for no in range(int(identifiers[0]), int(identifiers[1])+1):
        l = len(str(no))
        no_str = str(no)
        for s in range(1, l//2+1):
            if l % s == 0:
                pattern = no_str[:s]
                if no_str == pattern * (l // s):
                    result += no
                    break
        
print(result)
    
    
```
