```python
import numpy as np
```

```python
with open("05.txt", "r") as f:
    txt = f.readlines()
```

```python
# 1
first_part = True
test_ids = []
fresh_ranges = []
for entry in txt:
    if entry == '\n':
        first_part = False
    elif first_part:
        s, e = entry.split('-')
        fresh_ranges.append((int(s),int(e)+1))
    else:
        test_ids.append(int(entry))
```

```python
n_fresh_ids = 0
for idx in test_ids:
    for (s,e) in fresh_ranges:
        if idx >= s and idx <= e:
            n_fresh_ids += 1
            break
print(n_fresh_ids)
```

```python
# 2
added = False
all_ids = [fresh_ranges[0]]
for (s,e) in fresh_ranges[1:]:
    
    for i, (s_e, e_e) in enumerate(all_ids):
        if s < s_e and e > e_e:
            all_ids[i] = (s,e)
            added = True
            break
        elif s < s_e and e <= e_e and e >= s_e:
            all_ids[i] = (s,e_e)
            added = True
            break
        elif s >= s_e and s <= e_e and e > e_e:
            all_ids[i] = (s_e,e)
            added = True
            break
    if not added:
        all_ids.append((s,e))
```
