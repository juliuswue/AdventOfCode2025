```python
import numpy as np
from scipy import signal
```

```python
with open("04.txt", "r") as f:
    txt = f.readlines()
locs = [x.replace('\n', '') for x in txt]
```

```python
# 1
locs = [l.replace('.', '0').replace('@', '1') for l in locs]
locs = np.array([[int(l) for l in s] for s in locs])
kernel = np.array([[1,1,1],
                   [1,100,1],
                   [1,1,1]])

filtered_locs = signal.convolve2d(locs, kernel, mode='same')
print(np.sum((filtered_locs >= 100 ) & (filtered_locs < 104)))
```

```python
n_rolls = 0

while(True):
    filtered_locs = signal.convolve2d(locs, kernel, mode='same')
    acc_rolls = (filtered_locs >= 100 ) & (filtered_locs < 104)
    n_new_rolls = np.sum(acc_rolls)
    
    if n_new_rolls == 0:
        print(n_rolls)
        break
    else:
        locs[acc_rolls] = 0
        n_rolls += n_new_rolls
```
